# Linear Regression

```python
# gradient descent to minimize the mse

import numpy as np 

#Sample data
x = np.array([[1,2], [1,4], [1,6]]) # adding 1s for bias term(intercept)

y = np.array([3, 5, 7])

#weight initialization(2 features including intercept)
w = np.zeros(x.shape[1])
learning_rate=.01
n_iterations=2200
m=len(y)

for _ in range(n_iterations):
    predictions=x.dot(w)
    errors=predictions-y
    gradient=(2/m)* x.T.dot(errors)
    w=w-learning_rate*gradient

print("Weights after GD:", w)
print("Predictions:", x.dot(w))
```

# Stochastic Gradient Descent

```python
import numpy as np

X = np.array([2.0, 4.0, 6.0, 8.0])
y = np.array([4.0, 8.0, 12.0, 16.0])

w=0.0
b=0.0
learning_rate=.01
n_epochs=1000

for epoch in range(n_epochs):
    for i in range(len(X)):
        xi=X[i]
        yi=y[i]

        y_hat=w*xi+b 

        dw=2*(y_hat-yi)*xi
        db=2*(y_hat-yi)

        w-=learning_rate*dw
        b-=learning_rate*db

    if (epoch+1) % 25 ==0:
        print(f"Epoch {epoch+1}: w={w:.3f}, b={b:.3f}")

print("Final parameters: w=", w, ", b=", b)
```
