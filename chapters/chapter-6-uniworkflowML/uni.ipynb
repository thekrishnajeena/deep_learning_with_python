## Prepare the data

basics that are common to all data domains.

### Vectorization
all inputs and targets in a NN must be tensors of floating-point data(in specific cases, tensors of integers or strings), data vectorization

### Value Normalization
before feeding data into our network, had to normalize each feature independently so that it had a standard deviation of 1 and a mean of 0.

to make learning easier for the network, data should have the following characteristics: 

- take small values: typically in the 0-1 range
- be homogenous: all features should take values in roughly the same range.

not necessary(but can help)
- normalize each feature independently to have a mean of 0
- normalize each feature independently to have a standard deviation of 1.

with numpy arrays: 
- x -= x.mean(axis=0) #samples
- x /= x.std(axis=0)

### Handling missing values

- if the feature is categorical, safe to create a new category that means "the value is missing"
- if the feature is numerical, safe to replace the missing value with the average or median value for the feature in the dataset, could also train a model to predict the feature given the value of other features

### Choose an evaluation protocol
the goal of validaiton protocol is to accurately estimate what your success metric of choice(such as accuracy) will be on production data. the reliability of that process is critical to building a useful model.

### Beat a baseline
initial goal when working on the model is to achieve statistical power.

- feature engineering

- selecting a good enough training configuration
